```python
import pandas as pd
from vnstock3 import Quote

# Retrieve historical data as a DataFrame
quote = Quote('VNM')
df = quote.history(start='2020-01-01', end='2020-12-31')
df
```

```python
df.volume.div(1000_000).viz.boxplot(title='volume distribution', xlabel='VNM', ylabel='volume (million shares)', color_palette='stock', palette_shuffle=True)
```

```python
df.viz.combo(bar_data='volume', line_data='close', color_palette='stock', palette_shuffle=True, figsize=(12, 6))
```

```python
df.close.viz.timeseries()
```

```python
df.close.viz.hist(bins=20, color='skyblue', edgecolor='black')
```

```python
from vnstock3 import Company
company = Company('VNM')
shareholder_df = company.shareholders()
shareholder_df
```

```python
shareholder_df.viz.pie(labels='share_holder', values='share_own_percent', title='VNM shareholders', figsize=(8, 8))
```

```python
df.viz.pairplot()
```

```python
company.profile().T.viz.wordcloud(figsize=(12, 6))
```
